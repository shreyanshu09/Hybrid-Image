```
import cv2
import numpy as np
import scipy
from skimage import io
from matplotlib import pyplot as plt

def extract_high_frequency(image, a, b):
    high_frequency_image = image - extract_low_frequency(image, a)
    return high_frequency_image

def extract_low_frequency(image, b):
    low_frequency_image = cv2.GaussianBlur(image, (7, 7), b)
    return low_frequency_image

def hybrid_image_combine(high_frequency_image, low_frequency_image):
    hybridImage = high_frequency_image + low_frequency_image
    return hybridImage

a = 10
b = 10
```

```
image_1 = io.imread('/content/image/einstein.bmp')[:,:,0]
fig, axs = plt.subplots(1,5,figsize=(10,5))
axs[0].set_title('Image 1')
axs[0].imshow(image_1, cmap='gray')

high_frequency_image = extract_high_frequency(image, a, b)
axs[1].set_title('High Frequency Image')
axs[1].imshow(high_frequency_image, cmap='gray')

image_2 = io.imread('/content/image/marilyn.bmp')[:,:,0]
axs[2].set_title('Image 2')
axs[2].imshow(image_2, cmap='gray')

low_frequency_image = extract_low_frequency(image_2, a)
axs[3].set_title('Low Frequency Image')
axs[3].imshow(low_frequency_image, cmap='gray')

hybrid_image =  hybrid_image_combine(high_frequency_image, low_frequency_image)
axs[4].set_title('Hybrid Image')
axs[4].imshow(hybrid_image, cmap='gray')
```
